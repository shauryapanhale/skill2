```python
import pandas as pd

# Load the datasets
red_wine = pd.read_csv('/content/winequality-red.csv', sep=';')
white_wine = pd.read_csv('/content/winequality-white.csv', sep=';')

# Add a new column to differentiate red and white wine
red_wine['type'] = 'red'
white_wine['type'] = 'white'

# Merge the datasets
wine_data = pd.concat([red_wine, white_wine], axis=0, ignore_index=True)

# Display the first few rows of the merged dataset
print("Merged Dataset:")
print(wine_data.head())

# Save the merged dataset to a new CSV file
wine_data.to_csv('winequality-merged.csv', index=False)
print("Merged dataset saved as 'winequality-merged.csv'.")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder

# Load the merged dataset
wine_data = pd.read_csv('/content/winequality-merged.csv')

# Step 1: Check for missing values
print("Missing values per column:\n", wine_data.isnull().sum())

# Step 2: Handle missing values (if any)
# Here, we drop rows with missing values. You can also use imputation if needed.
wine_data = wine_data.dropna()

# Step 3: Encode categorical variable ('type')
label_encoder = LabelEncoder()
wine_data['type'] = label_encoder.fit_transform(wine_data['type'])  # 0 for red, 1 for white

# Step 4: Separate features (X) and target (y)
X = wine_data.drop('quality', axis=1)  # Features
y = wine_data['quality']  # Target (wine quality)

# Step 5: Normalize and scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 6: Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Print dataset shapes
print("Training set shape (X_train, y_train):", X_train.shape, y_train.shape)
print("Testing set shape (X_test, y_test):", X_test.shape, y_test.shape)
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

# Load the merged dataset
wine_data = pd.read_csv('/content/winequality-merged.csv')

# Step 1: Check for categorical columns
print("Dataset columns:\n", wine_data.dtypes)

# Step 2: Encode the 'type' column using different techniques

# 2.1: Label Encoding
label_encoder = LabelEncoder()
wine_data['type_label_encoded'] = label_encoder.fit_transform(wine_data['type'])  # 0 for red, 1 for white

# 2.2: One-Hot Encoding
wine_data = pd.get_dummies(wine_data, columns=['type'], prefix='type', drop_first=True)

# Display the updated dataset
print("Updated dataset with encoded columns:\n", wine_data.head())

# Save the preprocessed dataset
wine_data.to_csv('winequality-encoded.csv', index=False)
print("Dataset with encoding saved as 'winequality-encoded.csv'.")
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Load the preprocessed dataset
wine_data = pd.read_csv('winequality-encoded.csv')

# Select numerical columns for scaling
numerical_features = wine_data.select_dtypes(include=['float64', 'int64']).columns
numerical_features = numerical_features.drop('quality')  # Exclude target variable

# Step 1: Normalization (Min-Max Scaling)
minmax_scaler = MinMaxScaler()
wine_data[numerical_features + '_normalized'] = minmax_scaler.fit_transform(wine_data[numerical_features])

# Step 2: Standardization (Z-score Scaling)
standard_scaler = StandardScaler()
wine_data[numerical_features + '_scaled'] = standard_scaler.fit_transform(wine_data[numerical_features])

# Display the first few rows of the updated dataset
print("Dataset after normalization and scaling:\n", wine_data.head())

# Save the dataset with scaled features
wine_data.to_csv('winequality-normalized-scaled.csv', index=False)
print("Dataset saved as 'winequality-normalized-scaled.csv'.")
```

```python
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.preprocessing import MinMaxScaler

# Scale data to [0, 1] range for chi-square
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(wine_data.drop(columns=['quality']))
target = wine_data['quality']

# Apply chi-square
chi2_selector = SelectKBest(chi2, k=5)  # Select top 5 features
chi2_selector.fit(scaled_data, target)

# Selected features
chi2_features = wine_data.drop(columns=['quality']).columns[chi2_selector.get_support()]
print("Top features based on Chi-Square Test:\n", chi2_features)
```

```python
from sklearn.feature_selection import RFE
from sklearn.ensemble import RandomForestClassifier

# Define model
model = RandomForestClassifier(random_state=42)

# Recursive Feature Elimination
rfe = RFE(estimator=model, n_features_to_select=5)  # Select top 5 features
rfe.fit(wine_data.drop(columns=['quality']), wine_data['quality'])

# Selected features
rfe_features = wine_data.drop(columns=['quality']).columns[rfe.support_]
print("Top features based on RFE:\n", rfe_features)
```

```python
from sklearn.ensemble import RandomForestClassifier

# Train Random Forest
rf = RandomForestClassifier(random_state=42)
rf.fit(wine_data.drop(columns=['quality']), wine_data['quality'])

# Feature importance
importance = pd.DataFrame({'Feature': wine_data.drop(columns=['quality']).columns,
                           'Importance': rf.feature_importances_}).sort_values(by='Importance', ascending=False)

# Display top features
print("Top features based on Random Forest:\n", importance)

# Plot feature importance
sns.barplot(x='Importance', y='Feature', data=importance, palette='viridis')
plt.title('Feature Importance from Random Forest')
plt.show()
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Standardize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(wine_data.drop(columns=['quality']))

# Apply PCA
pca = PCA(n_components=5)  # Reduce to top 5 components
pca_data = pca.fit_transform(scaled_data)

# Explained variance
print("Explained variance by components:\n", pca.explained_variance_ratio_)

# Plot cumulative variance
plt.plot(range(1, 6), pca.explained_variance_ratio_.cumsum(), marker='o')
plt.title('Cumulative Explained Variance by PCA Components')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Variance')
plt.show()
```

```python
from sklearn.model_selection import train_test_split

# Features and target variable
X = wine_data.drop(columns=['quality'])
y = wine_data['quality']

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score

# Train Logistic Regression
lr = LogisticRegression(max_iter=1000)
lr.fit(X_train, y_train)

# Predictions
y_pred_lr = lr.predict(X_test)

# Evaluation
print("Logistic Regression Accuracy:", accuracy_score(y_test, y_pred_lr))
print(classification_report(y_test, y_pred_lr))
```

```python
from sklearn.tree import DecisionTreeClassifier

# Train Decision Tree
dt = DecisionTreeClassifier(random_state=42)
dt.fit(X_train, y_train)

# Predictions
y_pred_dt = dt.predict(X_test)

# Evaluation
print("Decision Tree Accuracy:", accuracy_score(y_test, y_pred_dt))
print(classification_report(y_test, y_pred_dt))
```

```python
from sklearn.ensemble import RandomForestClassifier

# Train Random Forest
rf = RandomForestClassifier(random_state=42)
rf.fit(X_train, y_train)

# Predictions
y_pred_rf = rf.predict(X_test)

# Evaluation
print("Random Forest Accuracy:", accuracy_score(y_test, y_pred_rf))
print(classification_report(y_test, y_pred_rf))
```

```python
from sklearn.svm import SVC

# Train SVM
svm = SVC(kernel='linear', random_state=42)
svm.fit(X_train, y_train)

# Predictions
y_pred_svm = svm.predict(X_test)

# Evaluation
print("SVM Accuracy:", accuracy_score(y_test, y_pred_svm))
print(classification_report(y_test, y_pred_svm))
```

```python
from sklearn.neighbors import KNeighborsClassifier

# Train k-NN
knn = KNeighborsClassifier(n_neighbors=5)  # You can experiment with different k values
knn.fit(X_train, y_train)

# Predictions
y_pred_knn = knn.predict(X_test)

# Evaluation
print("k-NN Accuracy:", accuracy_score(y_test, y_pred_knn))
print(classification_report(y_test, y_pred_knn))
```

```python
from sklearn.ensemble import GradientBoostingClassifier

# Train Gradient Boosting
gb = GradientBoostingClassifier(random_state=42)
gb.fit(X_train, y_train)

# Predictions
y_pred_gb = gb.predict(X_test)

# Evaluation
print("Gradient Boosting Accuracy:", accuracy_score(y_test, y_pred_gb))
print(classification_report(y_test, y_pred_gb))
```

```python
results = {
    "Logistic Regression": accuracy_score(y_test, y_pred_lr),
    "Decision Tree": accuracy_score(y_test, y_pred_dt),
    "Random Forest": accuracy_score(y_test, y_pred_rf),
    "SVM": accuracy_score(y_test, y_pred_svm),
    "k-NN": accuracy_score(y_test, y_pred_knn),
    "Gradient Boosting": accuracy_score(y_test, y_pred_gb)
}

# Display results
for classifier, accuracy in results.items():
    print(f"{classifier}: {accuracy:.4f}")
```
